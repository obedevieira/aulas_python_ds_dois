# Manipulação de dados

```python
import pandas as pd

url = "https://raw.githubusercontent.com/jfmiranda2/python_data_science/main/dataSets/Pokemon.csv"
df_pokemon = pd.read_csv(url)

#Resumo dos dados
df_pokemon.head(7) #Apresentar as primeiras linhas da tabela
df_pokemon.shape #Apresentar a quantidade de linhas e colunas da tabela
df_pokemon.describe() #Apresentas as estatísticas da tabela
```

```python
df_pokemon.columns = ['#', 'Nome', 'Tipo 1', 'Tipo 2', 'Total', 'HP', 'Ataque', 'Defesa',
       'Ataque Esp.', 'Defesa Esp.', 'Velocidade', 'Geração', 'Lendária']

df_pokemon.head(2)
```

```python

#renomeia as colunas#Leitura Colunas
df_pokemon['Ataque'].head() #Filtro de apenas uma coluna
#df_pokemon_filtrado = df_pokemon[['Ataque Esp.','Ataque', 'Defesa', 'Total']].head()
df_pokemon_filtrado = df_pokemon[['Ataque Esp.','Ataque', 'Defesa', 'Total']][13:20]
display(df_pokemon_filtrado)
```

```python
#Leitura Linhas
#.loc
df_pokemon_filtrado = df_pokemon[['Ataque Esp.','Ataque', 'Defesa', 'Total']][13:20]
df_pokemon_filtrado = df_pokemon.loc[13:20 ,['Ataque Esp.','Ataque', 'Defesa', 'Total']]
df_pokemon_filtrado
```

```python
#.iloc

df_pokemon.iloc[0:5, -1] # Filtro linha 0 até a linha 5 apresentando apenas a última coluna
df_pokemon.iloc[:, 0] #Todas as linhas com a primeira coluna
df_pokemon.iloc[[0, 7, 8], 5:8]# Filtro linhas específicas com as colunas 5 até a 8
```

```python
#Filtragem pelos dados

#AND = E = & - As duas condições devem ser verdadeiras para apresentar a linha
#df_pokemon.loc[(df_pokemon['Tipo 1'] == 'Grass') & (df_pokemon['Ataque'] > 100)]

#OR = OU = | = + - Uma das condições deve ser verdadeira para apresentar a linha
df_pokemon_filtrado = df_pokemon.loc[(df_pokemon['Tipo 1'] == 'Grass') | (df_pokemon['Tipo 1'] == 'Fire')]
df_pokemon_filtrado['Tipo 1'].value_counts()
```

```python
#Ordem das colunas

df_pokemon.sort_values(['Ataque', 'Defesa'], ascending= [True, False])
```

```python
df_pokemon_editado = df_pokemon.dropna(axis=1) #Remove as colunas apga as 300 e poucas linhas vazias
print(df_pokemon.shape)  #tabela original
print(df_pokemon_editado.shape)#tabela editada sem a coluna com dados vazios
display(df_pokemon) #apresenta tabela original
display(df_pokemon_editado)#apresenta tabela editada
```

```python
#Substituindo os valores vazios
df_pokemon_editado = df_pokemon.fillna('corinthias') #Substituir valores vazios com 0
df_pokemon_editado = df_pokemon.fillna(False) #Substituir valores vazios com False
print(df_pokemon.isnull().sum())
print(df_pokemon_editado.isnull().sum())
print(df_pokemon_editado['Tipo 2'].value_counts())
```

```python
#Substituindo os valores vazios
#df_pokemon_editado = df_pokemon.fillna(0) #Substituir valores vazios com 0
df_pokemon_editado = df_pokemon.fillna(method="ffill") #Substituir valores vazios o valor da linha anterior
print(df_pokemon.isnull().sum())
print(df_pokemon_editado.isnull().sum())
print(df_pokemon_editado['Tipo 2'].value_counts())
df_pokemon_editado.head(10)
```

# Operações e Transformações

```python
#Operações Matemáticas e Estatísticas Básicas
#mean(), sum(), std(), etc.
print(df_pokemon['Ataque'].max())
print(df_pokemon['Ataque'].min())
print(df_pokemon['Ataque'].mean())
print(df_pokemon['Ataque'].sum())
print(df_pokemon['Ataque'].std()) #desvio padrão https://www.todamateria.com.br/desvio-padrao/
```

```python
#Agrupamento de dados
df_pokemon.groupby(['Tipo 1']).sum()
#df_pokemon[['Tipo 1','Total', 'HP', 'Ataque', 'Defesa',
#       'Ataque Esp.', 'Defesa Esp.', 'Velocidade']].groupby(['Tipo 1']).sum()

df_pokemon.groupby('Tipo 1')[['Total','HP']].sum()
```
